```python
# Dependencies
import pandas as pd
import os

# Store filepath in a variable
file_one = os.path.join("clean_hawaii_measurements.csv")
file_two = os.path.join("clean_hawaii_stations.csv")

measure_df = pd.read_csv(file_one)
stations_df = pd.read_csv(file_two)
```

```python
measure_df['id'] = measure_df.index
stations_df['id'] = stations_df.index
stations_df.drop(['name'],axis=1)
clean_data1 = "measure_csv.csv"
measure_df.to_csv(clean_data1, index=False)
clean_data2 = "stations_csv.csv"
stations_df.to_csv(clean_data2, index=False)
```

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy import create_engine, MetaData
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric, Text, Float
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")
# Create a connection to the engine called `conn`

conn = engine.connect()
```

```python
Base = declarative_base()

class measurement1(Base):
    __tablename__ = 'Measurement1'
    
    id = Column(Integer, primary_key=True)
    date = Column(Text)
    station = Column(Text)
    prcp = Column(Float)
    tobs = Column(Integer)
    def __repr__(self):
        return f"id={self.id}, name={self.name}"
  

```

```python
class station3(Base):
    __tablename__ = 'Station3'
    
    id = Column(Integer, primary_key=True)
    station = Column(Text)
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Float)
    def __repr__(self):
        return f"id={self.id}, name={self.name}"
```

```python
Base.metadata.create_all(engine)
```

```python
df_of_data_to_insert = pd.read_csv(clean_data1)
data1 = df_of_data_to_insert.to_dict(orient='records')

data1[0]
```

```python
df_of_data_to_insert1 = pd.read_csv(clean_data2)
data2 = df_of_data_to_insert1.to_dict(orient='records')
data2[0]
```

```python
metadata = MetaData(bind=engine)
metadata.reflect()
```

```python
table1 = sqlalchemy.Table('Measurement1', metadata, autoload=True)
table2 = sqlalchemy.Table('Station3', metadata, autoload=True)
```

```python
conn.execute(table1.delete())
```

```python
conn.execute(table2.delete())
```

```python
conn.execute(table1.insert(), data1)
conn.execute(table2.insert(), data2)
```
